```python
name = "All 13 questions Internal vs External validation performance"
# Load data
path = '/jupyter/notebooks/biman/Paper/nature_paper/NM_code/data/'
data_path_v41_test_all = os.path.join(path, "whisper_large_v2_32_v41_test_all.pt")
data_v41_test_all = torch.load(data_path_v41_test_all)
data_path_rwe_test_v5_160_all = os.path.join(path, "whisper_large_v2_32_rwe_test_v5_160_all.pt")
data_rwe_test_v5_160_all = torch.load(data_path_rwe_test_v5_160_all)


# Extract relevant data
f1 = data_v41_test_all['F1_pos'][0]
sensitivity = data_v41_test_all['Sensitivity'][0]
specificity = data_v41_test_all['Specificity'][0]
roc_auc = data_v41_test_all['Auc'][0]
fpr = np.concatenate(([0], data_v41_test_all['Fpr'][0]))
tpr = np.concatenate(([0], data_v41_test_all['Tpr'][0]))

f1_data_rwe_test_v5_160_all = data_rwe_test_v5_160_all['F1_pos'][0]
sensitivity_data_rwe_test_v5_160_all = data_rwe_test_v5_160_all['Sensitivity'][0]
specificity_data_rwe_test_v5_160_all = data_rwe_test_v5_160_all['Specificity'][0]
roc_auc_data_rwe_test_v5_160_all = data_rwe_test_v5_160_all['Auc'][0]
fpr_data_rwe_test_v5_160_all = np.concatenate(([0], data_rwe_test_v5_160_all['Fpr'][0]))
tpr_data_rwe_test_v5_160_all = np.concatenate(([0], data_rwe_test_v5_160_all['Tpr'][0]))


# Plot data
colors = {
    'v41_test': {'line': '#6AA84F', 'fill': '#B6D7A8'},
    'rwe_test': {'line': '#C27BA0', 'fill': '#EA9999'}
}


# Plot data
plt.figure()
plt.plot(fpr, tpr, color='#6AA84F', lw=2, label=f'v41_test\nF1={f1:.4f}\nSensitivity={sensitivity:.4f}\nSpecificity={specificity:.4f}\nAUC={roc_auc:.4f}')
#plt.fill_between(fpr, tpr, alpha=0.2, color='#B6D7A8')

plt.plot(fpr_data_rwe_test_v5_160_all, tpr_data_rwe_test_v5_160_all, color='#C27BA0', lw=2, label=f'rwe_test\nF1={f1_data_rwe_test_v5_160_all:.4f}\nSensitivity={sensitivity_data_rwe_test_v5_160_all:.4f}\nSpecificity={specificity_data_rwe_test_v5_160_all:.4f}\nAUC={roc_auc_data_rwe_test_v5_160_all:.4f}')
#plt.fill_between(fpr_data_rwe_test_v5_160_all, tpr_data_rwe_test_v5_160_all, alpha=0.2, color='#EA9999')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f"{name}")


# Add ROC curve legend
plt.legend(loc='lower right')

# Add detailed metrics legends in two separate boxes
metric_legend_v41 = plt.Line2D([0], [0], color=colors['v41_test']['line'], lw=2, label=f'v41_test\nSensitivity={sensitivity:.4f}\nSpecificity={specificity:.4f}\nAUC={roc_auc:.4f}')
metric_legend_rwe = plt.Line2D([0], [0], color=colors['rwe_test']['line'], lw=2, label=f'rwe_test\nSensitivity={sensitivity_data_rwe_test_v5_160_all:.4f}\nSpecificity={specificity_data_rwe_test_v5_160_all:.4f}\nAUC={roc_auc_data_rwe_test_v5_160_all:.4f}')

legend_v41 = plt.legend(handles=[metric_legend_v41], loc='upper right', bbox_to_anchor=(1, 0.55))
plt.gca().add_artist(legend_v41)
plt.legend(handles=[metric_legend_rwe], loc='upper right', bbox_to_anchor=(1, 0.3))


#plt.savefig(f'roc_curve/{name}.png', dpi=300, bbox_inches='tight')
plt.show()
plt.close()
```
